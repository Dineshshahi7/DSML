```python
Q1 = df.Age.quantile(q = 0.25)
Q3 = df.Age.quantile(q = 0.75)
```

```python
IQR = Q3 - Q1
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
```

```python
df[(df.Age > upper_limit) | (df.Age < lower_limit)]
```
