#### 1. Simple way 

```python
#filling missing age by it's mean age of whole column
mean_age = df['Age'].mean()
df["Age"].fillna(mean_age, axis = 0, inplace = True)
```

#### 2. Intermediate way:
Fill missing value in age column by mean age of different column value based on sex column 

```python
# filter data 
male_data = df[df['Sex'] == 'male']

female_data = df[df['Sex'] == 'female']
```

```python
male_data['Age'].mean(), female_data['Age'].mean()
```

```python
df['Age'] = df['Age'].fillna(df.groupby('Sex')['Age'].transform('mean'))
```

#### 3. Best way: 
Filling missing value in age column by mean age of different initializer in name column

```python
df['Title'] = df['Name'].str.extract(r',\s*([^\.]*)\s*\.', expand=False)
```

```python
df['Title'].value_counts()
```

```python
# This code fill the mean value to all column either null value or not null value
mean_age_by_initializer = df.groupby("Title")['Age'].transform('mean')
mean_age_by_initializer
df['Age'] = df['Age'].fillna(mean_age_by_initializer, inplace = True)
```

```python
df['Age'] = df['Age'].fillna(df.groupby("Title")['Age'].transform('mean'))
```

```python
df['Age'].isnull().sum()
```

```python
df.Age.plot.hist(bins = 20)
```

##### Instead of using a global mean, you use a group-specific mean, which improves accuracy, especially when Age is an important feature

```python
titles = df['Name'].str.split(',')
df['Title'] = titles.str[1].str.strip().str.split(' ').str[0]
df.head()
```

```python
def get_titles(name):
    first_name = name.split(",")[-1]
    title = first_name.split(".")[0]
    return title.strip("")
```

```python
get_titles("Braund, Mr. Owen Harris")
```

```python
df['Title'] = df.Name.apply(get_titles)
```

```python
df.Name.transform(get_titles)
```
