```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "acc": [1, 2, 3],
    "name": [0, 9, 8],
    "age": [2, 4, 6]
})
```

```python
df2 = pd.DataFrame({
    "acc": [5, 9, 10],
    "loan": [44, 55, 66]
})
```

```python
row_concat = pd.concat([df1, df2], axis=0)
```

```python
row_concat
```

```python
col_concat = pd.concat([df1, df2], axis=1)
```

```python
col_concat
```

```python
df1 = pd.DataFrame({
    "Key": ["A", "B", "C", "D"],
    "inv_all": [99, 55, 33, 22],
    "acc": [1, 3, 5, 7]
})
```

```python
df2 = pd.DataFrame({
    "Key": ["E", "B", "C", "Z"],
    "hel_ration": [66, 0.88, 63, 29],
    "acc": [ 11, 32, 53, 74]
})
```

```python
pd.merge(df1, df2, how = "inner", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "outer", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "left", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "right", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
df1.merge(df2, how = "right", on = "Key", suffixes = ("_df1", "_df2"))
```
