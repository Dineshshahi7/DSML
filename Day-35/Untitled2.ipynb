```python
df1 = pd.DataFrame({
    "Key": ["A", "B", "C", "D"],
    "inv_all": [99, 55, 33, 22],
    "acc": [1, 3, 5, 7]
})
```

```python
df2 = pd.DataFrame({
    "Key": ["E", "B", "C", "Z"],
    "hel_ration": [66, 0.88, 63, 29],
    "acc": [ 11, 32, 53, 74]
})
```

```python
pd.merge(df1, df2, how = "inner", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "outer", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "left", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
pd.merge(df1, df2, how = "right", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
df1.merge(df2, how = "right", on = "Key", suffixes = ("_df1", "_df2"))
```

```python
df1 = pd.DataFrame({
    "Key": ["A", "B", "C", "D"],
    "inv_all": [99, 55, 33, 22],
    "acc": [1, 3, 5, 7]
})
```

```python
df1 = df1.set_index("Key")
df1
```

```python
df2 = pd.DataFrame({
    "Key": ["E", "B", "C", "Z"],
    "hel_ration": [66, 0.88, 63, 29],
    "acc": [ 11, 32, 53, 74]
})
```

```python
df2 = df2.set_index("Key")
df2
```

```python
df1.join(df2, how="inner", lsuffix="_x")
```

```python
df2.join(df1, how="right", rsuffix="_2")
```
