#### Conclusion
- Setosa has distinctly smaller sepal lengths, making it easy to separate from the other two species.

### **Problem 3: Handling Missing Data and Imputation**  
1. Identify columns with **missing values**.

   - For **numerical columns**, replace missing values with the **median**.
   - For **categorical columns**, replace missing values with the **most frequent** value.
3. Justify your choice of imputation methods and check if any missing data remains.

```python
data.isna().sum()
```

```python
for col in numerical_columns:
    data[col] = data[col].fillna(data[col].median())
```

```python
for col1 in categorical_columns:
    data[col1] = data[col1].fillna(data[col1].mode()[0])
```

### **Problem 4: Numpy Array Operations**  
1. Convert the numerical columns (such as `LotArea`, `YearBuilt`, `GrLivArea`, `SalePrice`, etc.) into a **NumPy array**.
2. Perform the following NumPy operations on the array:
   - Compute the **mean** and **standard deviation** of each column.
   - Normalize the data using **min-max scaling** and **z-score normalization**.
   - Compute the **dot product** between two numerical columns of your choice.

```python
import numpy as np
```

```python
num_array = data[numerical_columns].to_numpy()
```

```python
mean_vals = np.mean(num_array, axis=0)
std_vals = np.std(num_array, axis=0)
```

```python
# Min-Max scaling
min_vals = np.min(num_array, axis=0)
max_vals = np.max(num_array, axis=0)

min_max_scaled = (num_array - min_vals) / (max_vals - min_vals)
print("Min-Max Scaled (first 5 rows):\n", min_max_scaled[:5])
```

```python
# Z-score
z_score_scaled = (num_array - mean_vals) / std_vals
print("Z-Score Scaled (first 5 rows):\n", z_score_scaled[:5])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Price'], bins=30, kde=True)
plt.title("Distribution of SalePrice", fontsize=14)
plt.xlabel("SalePrice")
plt.ylabel("Count")
plt.show()
```

#### Observation
- Looking at the shape of the histogram and the KDE line, the data appears to be somewhat uniformly distributed, but with a few notable peaks and valleys.
- The distribution is not a perfect bell curve (normal distribution). There are several local peaks, suggesting that some price ranges are more common than others

### 2. Box plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Condition', y='Price', data=data)
plt.title("SalePrice by Overall Condition", fontsize=14)
plt.xlabel("Overall Condition")
plt.ylabel("SalePrice")
plt.show()
```

#### Observation
- Based on these observations, there appears to be no significant difference in SalePrice based on Overall Condition (Excellent, Good, Fair, or Poor).
- The median, interquartile range, and overall spread of sale prices are all very similar for each category.
- This is a surprising finding, as one would typically expect properties in "Excellent" condition to sell for higher prices than those in "Poor" condition.

### 3. Scatter plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Area', y='Price', data=data, alpha=0.7)
plt.title("Living Area vs SalePrice", fontsize=14)
plt.xlabel("Above-ground Living Area (sqft)")
plt.ylabel("SalePrice")
plt.show()
```

#### Observation
- There is no visible pattern or correlation between the two variables. The points are scattered uniformly across the plot, rather than forming a trend (e.g., an upward or downward slope).
- Normally, you would expect a positive correlation—as living area increases, the sale price should also increase. However, this plot does not show that relationship. This could be due to a problem with the data, such as a lack of variance in SalePrice or an issue with the data itself.

### 4. Heatmap

```python
plt.figure(figsize=(10,8))
corr = data.select_dtypes(include=['int64','float64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features", fontsize=14)
plt.show()
```
